```python
import cv2
import numpy as np
import os
from mtcnn.mtcnn import MTCNN

vid=cv2.VideoCapture('song.mp4')
count=0
while vid.isOpened():
    ret, frame = vid.read()
    if not ret:break
    
    detector = MTCNN()
    faces = detector.detect_faces(frame)
    
    if len(faces)>0:
        count+=1
        #grabcut
        mask = np.zeros(frame.shape[:2], np.uint8)
        rect = (50, 50, 450, 290)
        bgd_model = np.zeros((1, 65), np.float64)
        fgd_model = np.zeros((1, 65), np.float64) 
        cv2.grabCut(frame, mask, rect, bgd_model, fgd_model, 5, cv2.GC_INIT_WITH_RECT)
        mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
        result = frame * mask2[:, :, np.newaxis]

        for (x, y, w, h) in faces:
            cv2.rectangle(result, (x, y), (x+w, y+h), (255, 0, 0), 2)
        cv2.imwrite(f'anyimage{count}',result)
        cv2.waitKey(0)

    # cv2.imshow('result',result)
                   
vid.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import os

vid = cv2.VideoCapture('song.mp4')

output_path = 'grabcut_results/'
os.makedirs(output_path, exist_ok=True)

while vid.isOpened():
    ret, frame = vid.read()
    if not ret:
        break

    # Haar Cascade classifier
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    img = frame.copy()
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.3, minNeighbors=5)

    for i, (x, y, w, h) in enumerate(faces):
        # Extract each detected face region
        face_roi = img[y:y+h, x:x+w]

        # Apply GrabCut to the face region
        mask = np.zeros_like(face_roi[:, :, 0])
        rect = (0, 0, w, h)
        bgd_model = np.zeros((1, 65), np.float64)
        fgd_model = np.zeros((1, 65), np.float64)
        cv2.grabCut(face_roi, mask, rect, bgd_model, fgd_model, 5, cv2.GC_INIT_WITH_RECT)
        mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
        result = face_roi * mask2[:, :, np.newaxis]

        # Save the GrabCut result
        output_filename = os.path.join(output_path, f'grabcut_result_{i}.jpg')
        cv2.imwrite(output_filename, result)

    # Display the result
    cv2.imshow('Result with GrabCut', img)

    # Press 'q' to exit the video
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

vid.release()
cv2.destroyAllWindows()
```

```python
#DETECTING EVRYTHING ON FACE LIKE NOSE,MOUTH,EYES
import cv2
from mtcnn.mtcnn import MTCNN
detector = MTCNN()
img = cv2.imread('cbn2.jpg')
faces = detector.detect_faces(img)
print(faces)
for face in faces:
    x, y, width, height = face['box']
    cv2.rectangle(img, (x, y), (x + width, y + height), (0, 255, 0), 4)
    points = face['keypoints']
    eyeLeft = (points['left_eye'][0] - 10, points['left_eye'][1] - 10, 20, 20)
    eyeRight = (points['right_eye'][0] - 10, points['right_eye'][1] - 10, 20, 20)
    nose = (points['nose'][0] - 15, points['nose'][1] - 15, 30, 30)
    mouth_left = (points['mouth_left'][0] - 15, points['mouth_left'][1] - 10, 30, 20)
    mouth_right = (points['mouth_right'][0] - 15, points['mouth_right'][1] - 10, 30, 20)

    cv2.rectangle(img, eyeLeft, (255, 0, 0), 2)
    cv2.rectangle(img, eyeRight, (255, 0, 0), 2)
    cv2.rectangle(img, nose, (0, 0, 255), 2)
    cv2.rectangle(img, mouth_left, (0, 255, 0), 2)
    cv2.rectangle(img, mouth_right, (0, 255, 0), 2)
cv2.imshow('Face Detection', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

import cv2
from mtcnn.mtcnn import MTCNN
detector = MTCNN()
img = cv2.imread('GROUP.jpg')
faces = detector.detect_faces(img)
#print(faces)
color=(255, 0, 0)
for face in faces:
    if face['confidence'] > 0.98 :
        x, y, width, height = face['box']
        cv2.rectangle(img, (x, y), (x + width, y + height), color, 4)
        points = face['keypoints']
        eyeLeft = (points['left_eye'][0]- 10, points['left_eye'][1]-10 , 20, 20)
        eyeRight = (points['right_eye'][0]-10, points['right_eye'][1]-10, 20, 20)
        nose = (points['nose'][0]-15, points['nose'][1]-15, 30, 30)
        mouth_left = (points['mouth_left'][0] - 15, points['mouth_left'][1] - 10, 30, 20)
        mouth_right = (points['mouth_right'][0] - 15, points['mouth_right'][1] - 10, 30, 20)

        cv2.rectangle(img, eyeLeft, color, 2)
        cv2.rectangle(img, eyeRight, color, 2)
        cv2.rectangle(img, nose, (0, 0, 255), 2)
        cv2.rectangle(img, mouth_left, (0, 255, 0), 2)
        cv2.rectangle(img, mouth_right, (0, 255, 0), 2
    else:
        print("Face detecting accuracy is low",face['confidence'] )  
cv2.imshow('Face Detection', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
